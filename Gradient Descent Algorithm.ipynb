```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import matplotlib.transforms as mtransforms

from random import randint

%matplotlib inline

from sklearn.datasets import make_regression
```

```python
X,y = make_regression(n_samples=100, n_features=1, noise=5.6)
X = X.reshape((X.shape[0],))

plt.scatter(X, y)
plt.show()
```

```python
def gradient_descent(m, b, alpha, beta):
    
    m = m - alpha * 2 * sum((m * X + b - y)*X)
    b = b - beta * 2 * sum(m * X + b - y)
    
    
    return m,b
```

```python
# y = mX + b
m = randint(0, 10)
b = randint(0, 10)

iterations = 1000

# Determine a learning rate to intercept and slope
alpha = 0.01
beta  = 0.01

ep = 0.001
converge = False

J = sum((m * X + b - y)**2)
for i in range(iterations):
    
    m, b = gradient_descent(m, b, alpha, beta)
    error = sum((m * X + b - y)**2)
    
    if (J - error) <= ep:
        converge = True
        break
    J = error

if converge:    
    print('Converge with slope {} and intercept {}'.format(m,b))
else:
    print('Finished iterations with slope {} and intercept {}'.format(m,b))
```

```python
# Applying line to scatter plot
fig, ax = plt.subplots()
ax.scatter(X, y, c='black')
axes = plt.gca()
y_line = b + m * X
plt.plot(X, y_line, '--')
plt.show()
```
